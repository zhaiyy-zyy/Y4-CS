```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

### Step 1: Import Data 

The "real_estate_valuation_data_set.csv" file saves the market historical data of real estate valuation of a Chinese city. The variable information is as follows 

| Variable Name | Role | Type | Description | Units | 
|:--|:--|:--|:--|:--|
|No|ID|Integer||||
|X1 transaction date|Feature|Continuous|for example, 2013.250=2013 March, <br>2013.500=2013 June, etc.||
|X2 house age|Feature|Continuous||year|
|X3 distance to the <br>nearest metro station|Feature|Continuous||meter|
|X4 number of <br>convenience stores|Feature|Integer|number of convenience stores in the<br> living circle on foot|integer|
|X5 latitude|Feature|Continuous|geographic coordinate, latitude|degree|
|X6 longitude|Feature|Continuous|geographic coordinate, longitude|degree|
|Y house price of<br> unit area|Target|Continuous|10000 Chinese Yuan per <br>square metre|10000 CNY/<br>square metre|

```python
# Read the data with pandas
real_estate = pd.read_csv("real_estate_valuation_data_set.csv", index_col=0)
```

```python
# Show the first five rows of the data
real_estate.head()
```

```python
# Get the number of instances and features
row_num = real_estate.shape[0]
col_num = real_estate.shape[1]
```

```python
# Get the input variables X with features from X1 to X6.
X = np.array(real_estate.iloc[:,:6].values)
```

```python
# Get the target values y
y = np.array(real_estate.iloc[:,6].values)
```

We can transform the continuous target variable (house price) into the binary variable (house price range) according to the following rule.
|price range|category|label|
|:--:|:--:|:--:|
|$y\leq 2.5$|low|0|
|$y>2.5$|high|1|

```python
y_ = np.zeros(row_num, dtype="int")
y_[y<=2.5] = 0
y_[y>2.5] = 1
y_
```

### Step 2: Data Preprocessing

The values of features X1-X6 are in different scales. The feature in a large scale would have a huge influence on label prediction. We shall normalize the values of different features into a uniform scale to balance the contributions of different features to label prediction. Here we use the [MinMaxScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html) to scale the feature values into the range of $[0,1]$. Another choice for feature normalization is [StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html).

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

```python
X
```

### Step 3: Train-Test Split

We randomly split the data $X$ and $y$ into training and test sets with the ratio $1:1$.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_, test_size=0.5, random_state=0)
```

### Step 4: House Price Range Prediction with Gaussian Process Classification

Read [Scikit-Learn's GaussianProcessClassifier API](https://scikit-learn.org/stable/modules/generated/sklearn.gaussian_process.GaussianProcessClassifier.html), and implement the Gaussian process classification model with RBF kernel $k_{\mathrm{SE}}\left(\mathbf{x}_{i}, \mathbf{x}_{j}\right)=\sigma_f^{2} \exp \left(-\frac{1}{2 \ell^{2}}\left\|\mathbf{x}_{i}-\mathbf{x}_{j}\right\|_2^{2}\right)$.

**Task 1**: Implement the Gaussian process classification model with the hyperparameter setting: signal variance $\sigma_f^{2}=1.5$, length-scale $\ell=1.5$. Keep the hyperparameters fixed when fitting the model. 

```python
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF, ConstantKernel
# Train the Gaussian process classifier with the training data (X_train and y_train)
##: Write your code here
kernel = ConstantKernel(constant_value=1.5) * RBF(length_scale=1.5)

gpc = GaussianProcessClassifier(
    kernel=kernel,
    optimizer=None,
    random_state=0
)

gpc.fit(X_train, y_train)
```

```python
# Predict the labels of test data
##: Write your code 
y_pred = gpc.predict(X_test)
y_pred
```

```python
# Compare the ground-truth and predicted labels
##: Write your code here
pd.DataFrame({
    "Ground Truth": y_test,
    "Predicted": y_pred
})
```

Compute the accuracy score as an evaluation of the trained GaussianProcessClassifier model.

```python
##: Write your code here
acc = gpc.score(X_test, y_test)
print("Accuracy:", acc)
```

```python
# Print the log marginal likehood value of the fitted model
##: Write your code here
lml_value = gpc.log_marginal_likelihood_value_
print("Log Marginal Likelihood:", gpc.log_marginal_likelihood_value_)
```

**Task 2**: Use [GaussianProcessClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.gaussian_process.GaussianProcessClassifier.html)'s log marginal likelihood value to select the best hyperparameter setting when signal variance $\sigma_f^2$ varies in the set $\{1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5\}$, and length-scale $\ell$ varies in $\{1, 2, 3, 4, 5\}$. Write a grid search program to do model selection, do not use [GaussianProcessClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.gaussian_process.GaussianProcessClassifier.html)'s automatic hyperparameter optimizer, report the best hyperparameter setting as well as the corresponding model's prediction performance. 

```python
# Use grid search to find the best hyperparameter setting
##: Write your code here
sigma_f2_values = [1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5]
length_scale_values = [1, 2, 3, 4, 5]

best_lml = -np.inf
best_model = None
best_params = None

for sf in sigma_f2_values:
    for l in length_scale_values:

        kernel = ConstantKernel(sf) * RBF(l)

        gpc_model = GaussianProcessClassifier(
            kernel=kernel,
            optimizer=None,
            random_state=0
        )

        gpc_model.fit(X_train, y_train)

        lml = gpc_model.log_marginal_likelihood_value_

        if lml > best_lml:
            best_lml = lml
            best_model = gpc_model
            best_params = {"sigma_f2": sf, "length_scale": l}
```

```python
# Predict the labels of test data
##: Write your code here
y_pred_best = best_model.predict(X_test)
y_pred_best
```

```python
# Compare the ground-truth and predicted labels
##: Write your code here
pd.DataFrame({
    "Ground Truth": y_test,
    "Predicted": y_pred_best
})
```

Compute the accuracy score as an evaluation of the selected best GaussianProcessClassifier model.

```python
##: Write your code here
acc_best =best_model.score(X_test, y_test)
print("Best Accuracy:", acc_best)
```

```python
# Print the best model's log marginal likehood value
##: Write your code here
lml = best_model.log_marginal_likelihood_value_
print("Best Log Marginal Likelihood:", lml)
```

```python
# Print the best model's hyperparameter setting
##: Write your code here
print("Best hyperparameters:", best_params)
```

**Futher Exploration (Optional)**:
+ Choose [DotProduct](https://scikit-learn.org/stable/modules/generated/sklearn.gaussian_process.kernels.DotProduct.html) as the kernel function, evaluate the model's performance with the default kernel hyperparameter setting, and use grid search and [GaussianProcessClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.gaussian_process.GaussianProcessClassifier.html)'s hyperparameter optimizer to find the best hyperparameter setting. 

```python
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import DotProduct

kernel = DotProduct()

gpc_dot = GaussianProcessClassifier(
    kernel=kernel,
    random_state=0
)

gpc_dot.fit(X_train, y_train)

# predict
y_pred_dot = gpc_dot.predict(X_test)

# Accuracy
print("DotProduct (default) Accuracy:", gpc_dot.score(X_test, y_test))

# LML
print("DotProduct LML:", gpc_dot.log_marginal_likelihood_value_)
```

```python
sigma0_values = [0.1, 0.5, 1, 2, 5]

best_lml = -np.inf
best_model_dot = None
best_params_dot = None

for s in sigma0_values:

    kernel = DotProduct(sigma_0=s)

    model = GaussianProcessClassifier(
        kernel=kernel,
        optimizer=None,
        random_state=0
    )

    model.fit(X_train, y_train)

    lml = model.log_marginal_likelihood_value_

    if lml > best_lml:
        best_lml = lml
        best_model_dot = model
        best_params_dot = {"sigma_0": s}
        
print("Best DotProduct params:", best_params_dot)
print("Best DotProduct Accuracy:", best_model_dot.score(X_test, y_test))
print("Best DotProduct LML:", best_lml)
```

```python
kernel = DotProduct(
    sigma_0=1.0,
    sigma_0_bounds=(0.1, 5)
)

gpc_dot_opt = GaussianProcessClassifier(
    kernel=kernel,
    random_state=0
)

gpc_dot_opt.fit(X_train, y_train)

# predict
y_pred_opt = gpc_dot_opt.predict(X_test)

# Accuracy
print("Optimized DotProduct Accuracy:", gpc_dot_opt.score(X_test, y_test))

# LML
print("Optimized DotProduct LML:", gpc_dot_opt.log_marginal_likelihood_value_)

print("Optimized sigma_0:", gpc_dot_opt.kernel_.sigma_0)
```
