### Step 4: House Price Range Prediction with Categorical Naive Bayes

**Task 1**: Implement the Categorical Naive Bayes Classifier to predict house price range for the test house samples. **Do not** use Sciket-Learn's Naive Bayes implementation.  

```python
# Collect the feature value and label's co-occurrence counts
##: Write your code here
from collections import defaultdict

feature_counts = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))
label_counts = defaultdict(int)

for i in range(len(X_train)):
    label = y_train[i]
    label_counts[label] += 1

    for j in range(X_train.shape[1]):
        feature_value = X_train[i][j]
        feature_counts[j][feature_value][label] += 1
```

```python
# Predict the labels of test data with the collected count numbers, using Laplace smoothing with parameter alpha = 1
##: Write your code here
def predict_nb(X_test, feature_counts, label_counts, alpha=1):
    y_pred = []

    total_samples = sum(label_counts.values())
    labels = list(label_counts.keys())

    for x in X_test:
        probs = {}

        for label in labels:
            
            prob = np.log(label_counts[label] / total_samples)

            for j in range(len(x)):
                value = x[j]

                count = feature_counts[j][value][label]
                total = label_counts[label]

                prob += np.log((count + alpha) / (total + alpha * feat_val_nums[j]))

            probs[label] = prob

        y_pred.append(max(probs, key=probs.get))

    return np.array(y_pred)

y_pred = predict_nb(X_test, feature_counts, label_counts)
```

```python
# Compare the ground-truth and predicted labels
##: Write your code here
pd.DataFrame({"Ground Truth": y_test, "Predicted": y_pred})
```

```python
# Compute the accuracy score as an evaluation of the trained Naive Bayes Classifier
##: Write your code here
acc = (y_test == y_pred).mean()
print("Accuracy:", acc)
```
