| Variable Name | Role | Type | Description | Units |

We can transform the continuous target variable (house price) into the binary variable (house price range) according to the following rule.
|price range|category|label|
|:--:|:--:|:--:|
|$y\leq 2.5$|low|0|
|$y>2.5$|high|1|

```python
y_ = np.zeros(row_num, dtype="int")
y_[y<=2.5] = 0
y_[y>2.5] = 1
y_
```

### Step 2: Data Preprocessing

The values of features X1-X6 are in different scales. The feature in a large scale would have a huge influence on label prediction. We shall normalize the values of different features into a uniform scale to balance their contributions to label prediction. Here we use the [MinMaxScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html) to scale the feature values into the range of $[0,1]$. Another choice for feature normalization is [StandardScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html).

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)
```

### Step 3: Train-Test Split

We randomly split the data X and y_ into training and test sets with the ratio 1:1.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y_, test_size=0.5, random_state=0)
```

**Task 5**: [Scikit-Learn's SVC](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html) supports only a limited number of kernel choices. Sometimes we have to customize a kernel function. Read the section of "1.4.6.2. Custom Kernels" in the [kernel documentation](https://scikit-learn.org/stable/modules/svm.html#svm-kernels) and customize the linear-RBF combined kernel $k(\mathbf{x},\mathbf{x}^{\prime})=\left\langle\mathbf{x},\mathbf{x}^{\prime} \right\rangle+\exp\left(-\Vert\mathbf{x}-\mathbf{x}^{\prime}\Vert^2\right)$ with a python function. Pass the custom kernel function into the [SVC classifier](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html) and complete the binary classification task.  

```python
# Define the customized kernel function
def combined_kernel(X1, X2):
    # X1 is a matrix of shape (n_samples_1, n_features)
    # X2 is a matrix of shape (n_samples_2, n_features)
    # the function should return a kernel matrix of shape (n_samples_1, n_samples_2)
    ##: Write your code here
    linear_part = np.dot(X1, X2.T)

    X1_sq = np.sum(X1**2, axis=1).reshape(-1, 1)
    X2_sq = np.sum(X2**2, axis=1).reshape(1, -1)

    dist_sq = X1_sq + X2_sq - 2 * np.dot(X1, X2.T)

    rbf_part = np.exp(-dist_sq)

    return linear_part + rbf_part
```

```python
from sklearn.svm import SVC
# Train the custom kernel based support vector machine with the training data (X_train and y_train)
##: Write your code here
svm_custom = SVC(kernel=combined_kernel)
svm_custom.fit(X_train, y_train)
```

```python
# Predict the labels of test data
##: Write your code here
y_pred_custom = svm_custom.predict(X_test)
```

```python
# Compare the ground-truth and predicted labels
##: Write your code here
pd.DataFrame({
    "Ground Truth": y_test,
    "Predicted": y_pred_custom
})
```

```python
# Compute the accuracy score as an evaluation of the custom kernel based support vector machine
##: Write your code here
acc_custom = svm_custom.score(X_test, y_test)
print("Custom Kernel SVM Accuracy:", acc_custom)
```

**Futher Exploration (Optional)**:
+ Read [Scikit-Learn's LinearSVR](https://scikit-learn.org/stable/modules/generated/sklearn.svm.LinearSVR.html) and [SVR](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html) API, complete the house price prediction (**regression**) task using the Support Vector Regression model with the linear, Polynomial, RBF, Sigmoid and custom kernel functions. 

```python
from sklearn.svm import LinearSVR
from sklearn.metrics import mean_squared_error, r2_score

y_reg = real_estate.iloc[:, 6].to_numpy()

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.5, random_state=0
)
scaler_reg = MinMaxScaler()
X_train_reg = scaler_reg.fit_transform(X_train_reg)
X_test_reg = scaler_reg.transform(X_test_reg)

svr_linear = LinearSVR(max_iter=10000, random_state=0)
svr_linear.fit(X_train_reg, y_train_reg)

y_pred_linear = svr_linear.predict(X_test_reg)

mse_linear = mean_squared_error(y_test_reg, y_pred_linear)
r2_linear = r2_score(y_test_reg, y_pred_linear)
```

```python
from sklearn.svm import SVR

svr_poly = SVR(kernel='poly', degree=3, coef0=0.5, gamma=0.75)
svr_poly.fit(X_train_reg, y_train_reg)

y_pred_poly = svr_poly.predict(X_test_reg)

mse_poly = mean_squared_error(y_test_reg, y_pred_poly)
r2_poly = r2_score(y_test_reg, y_pred_poly)
```

```python
svr_rbf = SVR(kernel='rbf', gamma=0.5)
svr_rbf.fit(X_train_reg, y_train_reg)
y_pred_rbf = svr_rbf.predict(X_test_reg)

mse_rbf = mean_squared_error(y_test_reg, y_pred_rbf)
r2_rbf = r2_score(y_test_reg, y_pred_rbf)
```

```python
svr_sigmoid = SVR(kernel='sigmoid', gamma=0.25, coef0=0.5)
svr_sigmoid.fit(X_train_reg, y_train_reg)

y_pred_sigmoid = svr_sigmoid.predict(X_test_reg)

mse_sigmoid = mean_squared_error(y_test_reg, y_pred_sigmoid)
r2_sigmoid = r2_score(y_test_reg, y_pred_sigmoid)
```

```python
svr_custom = SVR(
    kernel=lambda X1, X2: (
        np.dot(X1, X2.T) +
        np.exp(
            -(np.sum(X1**2, axis=1).reshape(-1, 1)
            + np.sum(X2**2, axis=1).reshape(1, -1)
            - 2 * np.dot(X1, X2.T))
        )
    )
)

svr_custom.fit(X_train_reg, y_train_reg)

y_pred_custom = svr_custom.predict(X_test_reg)

mse_custom = mean_squared_error(y_test_reg, y_pred_custom)
r2_custom = r2_score(y_test_reg, y_pred_custom)
```

```python
results = pd.DataFrame({
    "Model": ["Linear", "Poly", "RBF", "Sigmoid", "Custom"],
    "MSE": [mse_linear, mse_poly, mse_rbf, mse_sigmoid, mse_custom],
    "R2": [r2_linear, r2_poly, r2_rbf, r2_sigmoid, r2_custom]
})

print(results)
```
