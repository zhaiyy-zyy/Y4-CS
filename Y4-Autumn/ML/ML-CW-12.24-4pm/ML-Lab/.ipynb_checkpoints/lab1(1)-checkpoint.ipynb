```python
from sklearn.datasets import fetch_openml
import numpy as np

print("Dowmloading MNIST dataset...")

X, y = fetch_openml('mnist_784', data_home='./', return_X_y=True, as_frame=False)

X = X.astype(np.float32) / 255.0

# 3. 定义 x1, x2：取前两张图片并 reshape 成 28×28
x1 = X[0].reshape(28, 28)
x2 = X[1].reshape(28, 28)
x3 = X[2].reshape(28, 28)
x4 = X[3].reshape(28, 28)
x5 = X[4].reshape(28, 28)
x6 = X[5].reshape(28, 28)
x7 = X[6].reshape(28, 28)
x8 = X[7].reshape(28, 28)
x9 = X[8].reshape(28, 28)
x10 = X[9].reshape(28, 28)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 3))               # 新建一张空白画布

plt.subplot(2, 5, 1)        # 把画布分成 1 行 2 列，这里画第 1 张
plt.imshow(x1, cmap='gray') # 显示第一张图，用灰度显示
plt.xticks([]), plt.yticks([])  # 去掉横纵坐标的刻度

plt.subplot(2, 5, 2)        # 画第 2 张图
plt.imshow(x2, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 5, 3)        # 画第 3 张图
plt.imshow(x3, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 5, 4)        # 画第 4 张图
plt.imshow(x4, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 5, 5)        # 画第 5 张图
plt.imshow(x5, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 5, 6)        # 画第 6 张图
plt.imshow(x6, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 5, 7)        # 画第 7 张图
plt.imshow(x7, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 5, 8)        # 画第 8 张图
plt.imshow(x8, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 5, 9)        # 画第 9 张图
plt.imshow(x9, cmap='gray')
plt.xticks([]), plt.yticks([])

plt.subplot(2, 5, 10)        # 画第 10 张图
plt.imshow(x10, cmap='gray')
plt.xticks([]), plt.yticks([])

#plt.tight_layout() #自动调整间距
plt.show()                  # 显示整个窗口
```
