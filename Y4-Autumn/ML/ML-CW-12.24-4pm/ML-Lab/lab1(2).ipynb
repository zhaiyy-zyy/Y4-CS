```python
from sklearn.datasets import fetch_openml
import numpy as np
import matplotlib.pyplot as plt

print("Dowmloading MNIST dataset...") #下载数据集

X, y = fetch_openml('mnist_784', data_home='./', return_X_y=True, as_frame=False) #打开数据集使用frame

X = X.astype(np.float32) / 255.0 #归一化像素值，除以 255 后，范围变成 0 – 1 之间的小数，能让模型训练更稳定
```

```python
# 2. 绘制前 10 张手写数字
plt.figure(figsize=(8, 3)) #指定画布的宽度为 8 英寸、高度为 3 英寸

# range(10) 表示从 0 到 9。每次循环都绘制一张图片。
for i in range(10):
    plt.subplot(2, 5, i + 1) #把整张画布分成 2 行 5 列的格子，一共 10 个小子图。
    plt.imshow(X[i].reshape(28, 28), cmap='gray')
    plt.axis('off')  # 去掉坐标轴
plt.tight_layout() #自动调整间距
plt.show()
```
